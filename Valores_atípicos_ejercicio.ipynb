```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cop=pd.read_csv('Ventas_totales.csv')
cop.head()
```

```python
valores_nulos=cop.isnull().sum()
valores_nulos
```

```python
df=cop.copy()
df
```

```python
cop['salon_ventas']=cop['salon_ventas'].fillna(round(cop['salon_ventas'].mean(),1))
cop['tarjetas_debito']=cop['tarjetas_debito'].fillna(round(cop['tarjetas_debito'].median(),2))
cop['tarjetas_credito']=cop['tarjetas_credito'].fillna(0)
cop['otros_medios']=cop['otros_medios'].fillna(1)
cop['subtotal_ventas_alimentos_bebidas']=cop['subtotal_ventas_alimentos_bebidas'].fillna(method='ffill')
cop['bebidas']=cop['bebidas'].fillna(0)
cop['almacen']=cop['almacen'].fillna(0)
cop['panaderia']=cop['panaderia'].fillna(0)
cop['lacteos']=cop['lacteos'].fillna(0)
cop['carnes']=cop['carnes'].fillna(0)
cop['verduleria_fruteria']=cop['verduleria_fruteria'].fillna(method='ffill')
cop['alimentos_preparados_rotiseria']=cop['alimentos_preparados_rotiseria'].fillna(method='ffill')
cop['indumentaria_calzado_textiles_hogar']=cop['indumentaria_calzado_textiles_hogar'].fillna(method='bfill')
cop['electronicos_articulos_hogar']=cop['electronicos_articulos_hogar'].fillna(method='bfill')
cop['otros']=cop['otros'].fillna(method='bfill')
```

```python
valores_nulos=cop.isnull().sum()
valores_nulos
```

**VENTAS_PRECIOS_CORRIENTES**

**Método desviación estandar**

```python
col1=df["ventas_precios_corrientes"]
Limite_Superior= col1.mean() + 3*col1.std()
Limite_Inferior= col1.mean() - 3*col1.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)

outliers= df[(col1>Limite_Superior)|(col1<Limite_Inferior)]
display (outliers)

data_clean= cop[(col1<=Limite_Superior)&(col1>=Limite_Inferior)]
display (data_clean)

fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean["ventas_precios_corrientes"]) 
plt.title("Outliers")
plt.show()
```

**Método rango intercuartílico**

```python
col1=cop["ventas_precios_corrientes"]

percentile25=col1.quantile(0.25) #Q1
percentile75=col1.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)

outliers_iqr= cop[(col1>Limite_Superior_iqr)|(col1<Limite_Inferior_iqr)]
display(outliers_iqr)

data_clean_iqr= cop[(col1<=Limite_Superior_iqr)&(col1>=Limite_Inferior_iqr)]
display (data_clean_iqr)

fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean["ventas_precios_corrientes"]) 
plt.title("Outliers")
plt.show()
```
